```python
# Notebook: notebooks/LSTM_sentiment_classifier_TestClean_Final.ipynb
print("Init: Starting LSTM Sentiment Classifier Notebook (using test_clean, integrated logic)...")
from tensorflow.keras import layers
import random

# --- Título del Notebook ---
# # Clasificación de Sentimiento de Tweets con LSTM y GridSearchCV (Usando 'test_clean' y Lógica Integrada)

# --- Descripción ---
# Este notebook implementa, entrena y evalúa dos modelos de Red Neuronal LSTM
# para clasificar el sentimiento de tweets (positivo, negativo, neutral)
# del dataset del Mundial FIFA 2022, utilizando aprendizaje supervisado.
# Utiliza la columna 'test_clean' como entrada de texto, Bidirectional LSTM
# y sparse_categorical_crossentropy con etiquetas numéricas enteras.
# Se presenta un modelo LSTM con hiperparámetros fijos como baseline,
# y otro modelo LSTM optimizado usando búsqueda de hiperparámetros con GridSearchCV.
# Incluye preprocesamiento (carga de 'test_clean'), división de datos (train/val/test),
# entrenamiento de modelos para evaluación y para despliegue (usando todos los datos),
# evaluación estadística (métricas y matriz de confusión con anotaciones), visualización
# del historial de entrenamiento (para el modelo fijo) y guardado de ambos modelos finales.

# --- Objetivos Cubiertos (Unidad 3) ---
# - OE3.3: Comprensión de aprendizaje supervisado, PNL y redes neuronales (LSTM Bidireccional).
# - OE3.4: Aplicación de redes neuronales (LSTM) para resolver tareas de PNL (clasificación de sentimiento) usando programación.
# - Cumplimiento del requisito de "Find its best hyperparameters using functions such as GridSearchCV", aprovechando el paralelismo de CPU.
# - Evaluación detallada con métricas (Accuracy, Precision, Recall, F1-score, Kappa) y visualizaciones (historial, matriz de confusión).
# - Guardado de modelos entrenados para su uso futuro.

# --- Importaciones Necesarias ---
import pandas as pd
import numpy as np
import os
# import ast # No necesitamos ast.literal_eval ya que test_clean es string
import sys # Para salir si hay errores críticos
import warnings # Para manejar advertencias si es necesario
import matplotlib.pyplot as plt # Para gráficos
import seaborn as sns # Para gráficos más bonitos (especialmente matriz de confusión)


# Sklearn para split de datos y métricas
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, cohen_kappa_score
from sklearn.metrics import confusion_matrix # Para la matriz de confusión
from sklearn.metrics import classification_report # Para reporte completo de clasificación

# TensorFlow/Keras para la red neuronal
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Bidirectional, Dense, Dropout # Importar Bidirectional
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
# from tensorflow.keras.utils import to_categorical # No necesaria con sparse_categorical_crossentropy y etiquetas enteras
from tensorflow.keras.callbacks import History # Para capturar el historial en el modelo fijo

# Wrapper para usar modelos Keras con GridSearchCV de scikit-learn
from scikeras.wrappers import KerasClassifier

# Para la búsqueda de hiperparámetros en Keras/TensorFlow con Scikit-learn
from sklearn.model_selection import GridSearchCV

# Ignorar warnings de scikit-learn y KerasClassifier que a veces pueden aparecer
warnings.filterwarnings('ignore', category=UserWarning, module='sklearn')
warnings.filterwarnings('ignore', category=FutureWarning, module='sklearn')
warnings.filterwarnings('ignore', category=DeprecationWarning, module='scikeras')
warnings.filterwarnings('ignore', category=UserWarning, module='tensorflow')


# ================================================================
# 0. Silencio de logs de TensorFlow y Semilla para reproducibilidad
# ================================================================
os.environ["TF_CPP_MIN_LOG_LEVEL"]  = "3"   # 0=DEBUG,1=INFO,2=WARNING,3=ERROR
os.environ["TF_ENABLE_ONEDNN_OPTS"] = "0"   # (opcional) desactiva oneDNN

SEED = 42
random.seed(SEED); np.random.seed(SEED); tf.random.set_seed(SEED)

# --- Configuración de Hardware (Forzar uso de CPU) ---
print("Configurando TensorFlow para usar solo CPU.")
try:
    gpu_devices = tf.config.list_physical_devices('GPU')
    if gpu_devices:
        tf.config.set_visible_devices([], 'GPU')
        print(f"GPUs detectadas: {len(gpu_devices)}. Explicitamente deshabilitadas para usar solo CPU.")
    else:
        print("No se detectaron GPUs físicas. TensorFlow usará CPU por defecto.")
    print("Dispositivos físicos visibles para TensorFlow:", tf.config.list_physical_devices())
except Exception as e:
    print(f"Error al intentar configurar TensorFlow para solo CPU: {e}")
    print("Continuando. TensorFlow usará CPU por defecto si la configuración falló o no hay GPU.")


# --- Configuración de Rutas y Parámetros ---
input_folder = '../data_processed' # O la ruta correcta a tu archivo
input_filename = 'fifa_tweets_clean.csv' # Nombre del archivo CSV
input_csv_file = os.path.join(input_folder, input_filename)

# Definir la carpeta de salida para los modelos guardados
ruta_base_proyecto = os.path.dirname(os.getcwd())
models_output_folder = os.path.join(ruta_base_proyecto, 'models')
os.makedirs(models_output_folder, exist_ok=True)
print(f"Los modelos entrenados se guardarán en: {models_output_folder}")


# Definir las columnas a usar del archivo procesado
CLEAN_TEXT_INPUT_COLUMN = 'test_clean' # Columna con el texto limpio (string)
LABEL_COLUMN = 'sentiment_label'     # Columna con la etiqueta numérica del sentimiento (int 0, 1, 2)
ORIGINAL_SENTIMENT_COLUMN = 'Sentiment' # Columna original del sentimiento (para referencia/mapeo inverso)


# Hyperparámetros para la preparación de datos (Tokenización y Padding)
MAX_WORDS = 10000         # Número máximo de palabras únicas en el vocabulario del tokenizer (las más frecuentes). None para usar todo el vocabulario.
MAX_SEQUENCE_LENGTH = 50 # <-- Usar MAX_LEN=50 como en tu código que funcionó
EMBEDDING_DIM = 128       # <-- Usar embed_dim=128 como en tu código que funcionó

# Configuración de entrenamiento y evaluación
TEST_SIZE = 0.20          # Proporción del dataset para el conjunto de prueba (20%)
# VAL_SIZE no se usa en split explícito con validation_split en .fit, pero mantenemos la proporción para el split inicial si se necesitara en GridSearchCV cv
VAL_SPLIT_IN_FIT = 0.1    # Proporción del conjunto de entrenamiento (X_train) a usar para validación DENTRO de .fit()
RANDOM_STATE = SEED       # Para reproducibilidad

# --- Cargar el dataset procesado ---
print(f"\nCargando dataset procesado desde '{input_csv_file}'...")
try:
    df = pd.read_csv(
        input_csv_file,
        encoding='utf-8',
        # No necesitamos converters para 'processed_tokens' si no la usamos para X.
    )

    print("Dataset procesado cargado exitosamente.")
    print(f"Número total de filas cargadas: {len(df)}")
    print(f"Columnas cargadas: {df.columns.tolist()}")

except FileNotFoundError:
    print(f"Error: El archivo procesado '{input_csv_file}' no fue encontrado.")
    print(f"Asegúrate de que la ruta '{input_csv_file}' es correcta y que el archivo existe.")
    sys.exit(1)
except Exception as e:
    print(f"Error al cargar o procesar el archivo CSV: {e}")
    print("Asegúrate de que el archivo no esté corrupto.")
    sys.exit(1)

# --- Verificación y Preparación de Datos ---

# 1. Eliminar filas con etiquetas de sentimiento faltantes o no válidas
initial_rows = len(df)
df_cleaned = df[df[LABEL_COLUMN] != -1].copy()
print(f"\nFilas con sentiment_label = -1 eliminadas: {initial_rows - len(df_cleaned)}")

# 2. Eliminar filas donde la columna de texto de entrada es NaN o está vacía/solo espacios
# Convertir a string antes de verificar NaN por si acaso
df_cleaned[CLEAN_TEXT_INPUT_COLUMN] = df_cleaned[CLEAN_TEXT_INPUT_COLUMN].astype(str)
df_cleaned = df_cleaned[df_cleaned[CLEAN_TEXT_INPUT_COLUMN].apply(lambda x: x.strip() != "")].copy()
print(f"Filas con {CLEAN_TEXT_INPUT_COLUMN} vacío o NaN eliminadas: {initial_rows - len(df_cleaned) - (initial_rows - len(df_cleaned))}") # Corrección en el cálculo

# Convertir las etiquetas numéricas a tipo entero (verificación)
df_cleaned[LABEL_COLUMN] = df_cleaned[LABEL_COLUMN].astype(int)

print(f"\nNúmero final de filas para modelado: {len(df_cleaned)}")
# Verificar la distribución de las etiquetas numéricas válidas
print(f"Distribución de etiquetas ({LABEL_COLUMN}): \n{df_cleaned[LABEL_COLUMN].value_counts()}")

# Determinar el número de clases y el mapeo inverso
unique_labels = sorted(df_cleaned[LABEL_COLUMN].unique())
num_classes = len(unique_labels)
print(f"Número de clases detectado: {num_classes}")

# Mapeo inverso de etiquetas numéricas a nombres
reverse_sentiment_mapping = {0: 'negative', 1: 'neutral', 2: 'positive'}
valid_target_names = [reverse_sentiment_mapping[i] for i in unique_labels if i in reverse_sentiment_mapping]
if len(valid_target_names) != num_classes:
    print("Advertencia: Las etiquetas numéricas en los datos no coinciden completamente con el mapeo reverse_sentiment_mapping.")
    target_names = [f'Class_{i}' for i in unique_labels] # Crea nombres genéricos si el mapeo falla
else:
    target_names = valid_target_names

print(f"Mapeo de etiquetas: {reverse_sentiment_mapping}")
print(f"Nombres de clases para reporte: {target_names}")


# 3. Separar características (X) y etiquetas (y)
X = df_cleaned[CLEAN_TEXT_INPUT_COLUMN] # Serie de strings de 'test_clean'
y = df_cleaned[LABEL_COLUMN] # Serie/array de etiquetas numéricas enteras


# 4. Tokenización y construcción del vocabulario (con Keras Tokenizer)
print(f"\nRealizando tokenización y padding usando '{CLEAN_TEXT_INPUT_COLUMN}'...")
# Usar num_words=MAX_WORDS para limitar el vocabulario
tokenizer = Tokenizer(num_words=MAX_WORDS, oov_token="<OOV>") # oov_token para palabras fuera del vocabulario
# Ajustar el tokenizer DIRECTAMENTE en la Serie X (que contiene las strings de test_clean)
tokenizer.fit_on_texts(X)

# Convertir el texto a secuencias de enteros usando el vocabulario aprendido
X_sequences = tokenizer.texts_to_sequences(X)

# Imprimir tamaño del vocabulario real y efectivo para Embedding
# input_dim en Embedding debe ser 1 + máximo índice entero en los datos de entrada.
# Si num_words=N, los índices más altos serán <= N si no hay OOV, o N+1 si hay OOV.
# len(tokenizer.word_index) da el número de palabras únicas *vistas*, sin padding.
# Si num_words es None, vocab_size = len(word_index) + 1 (para padding).
# Si num_words=N, vocab_size = min(len(word_index), N) + 1 (para padding).
# PERO, si hay OOV, y num_words=N, OOV token gets index 1, top N words get 2..N+1. Padding is 0. Size is N+2.
# La forma más segura: obtener el tamaño del vocabulario del tokenizer + 1 (para padding).
# Si num_words está definido, el tokenizer mapeará solo hasta ese límite (+ OOV).
# El tamaño real de los embeddings necesarios es el máximo índice + 1.
# Keras Tokenizer word_index starts from 1. Max index is len(word_index). Add 1 for padding.
effective_vocab_size = len(tokenizer.word_index) + 1
if MAX_WORDS is not None and MAX_WORDS > 0 and MAX_WORDS < len(tokenizer.word_index):
    # Si num_words limitó el vocabulario, el tokenizer.word_index aún mapea todas las palabras vistas,
    # pero texts_to_sequences solo usa los índices de las top num_words (+OOV).
    # El índice OOV (si existe) suele ser 1. Los top num_words suelen ir de 2 a num_words+1. Padding es 0.
    # Total size es num_words + 2 si hay OOV, o num_words + 1 si no hay OOV.
    # Sin embargo, Scikeras wrapper con sparse_categorical_crossentropy parece esperar input_dim = len(word_index) + 1
    # si num_words is None, o MAX_WORDS + 1 si num_words is set (sin contar OOV index special case).
    # Vamos a usar len(tokenizer.word_index) + 1 que es el tamaño total del vocabulario que el tokenizer podría generar.
    # Si MAX_WORDS está seteado, el tokenizer.word_index contendrá más palabras, pero texts_to_sequences y Embedding respetarán MAX_WORDS.
    # Revertir a la lógica simple de len(tokenizer.word_index) + 1
    pass # effective_vocab_size ya está calculado

print(f"Tamaño del vocabulario (num_words en Tokenizer): {MAX_WORDS} (None para todo)")
print(f"Tamaño real del word_index del Tokenizer: {len(tokenizer.word_index)}")
print(f"Tamaño efectivo del vocabulario a usar en Embedding (incl. padding): {effective_vocab_size}")
print(f"Primeras 10 palabras del vocabulario (índice a palabra):")
index_word = {index: word for word, index in tokenizer.word_index.items()}
for i in range(1, min(11, len(index_word) + 1)):
     if i in index_word:
         print(f"  {i}: {index_word[i]}")


# 5. Padding de secuencias
# Asegurar que todas las secuencias tengan la misma longitud para TODO el dataset
X_padded = pad_sequences(X_sequences, maxlen=MAX_SEQUENCE_LENGTH, padding='post', truncating='post')

# Las etiquetas 'y' ya están como enteros, perfectas para sparse_categorical_crossentropy.
# No necesitamos y_one_hot para el entrenamiento, solo para stratify en train_test_split si sklearn lo requiriera con enteros.
# Para estratificar, sklearn.model_selection.train_test_split funciona bien con etiquetas enteras 1D.

# 6. Dividir los datos en conjuntos de entrenamiento y prueba (etiquetas enteras 1D)
# Nota: Aquí no creamos X_val/y_val explícitamente en el split inicial,
# usamos validation_split dentro del .fit() del modelo fijo y el retrain del mejor modelo.
# GridSearchCV usa cv folds para validación, no necesita X_val/y_val explícito aquí.
X_train, X_test, y_train, y_test = train_test_split(
    X_padded, # Usar X_padded (secuencias numéricas)
    y,        # Usar y (etiquetas enteras 1D)
    test_size=TEST_SIZE,
    random_state=RANDOM_STATE,
    stratify=y # Estratificar usando las etiquetas enteras 1D
)

print(f"\n--- Datos listos para el modelado ---")
print(f"Forma de los datos de entrenamiento (X_train): {X_train.shape}")
print(f"Forma de las etiquetas de entrenamiento (y_train - enteros): {y_train.shape}")
print(f"Forma de los datos de prueba (X_test): {X_test.shape}")
print(f"Forma de las etiquetas de prueba (y_test - enteros): {y_test.shape}")
print(f"Forma de los datos COMPLETOS (X_padded): {X_padded.shape}") # Usado para entrenamiento final
print(f"Forma de las etiquetas COMPLETAS (y - enteros): {y.shape}") # Usado para entrenamiento final


print(f"Longitud máxima de secuencia: {MAX_SEQUENCE_LENGTH}")
print(f"Dimensión de embedding: {EMBEDDING_DIM}")
print(f"Tamaño efectivo del vocabulario usado: {effective_vocab_size}")
print(f"Número de clases: {num_classes}")


# --- Definir la función constructora del modelo LSTM Bidireccional ---
# Usa sparse_categorical_crossentropy
def build_lstm_model(vocab_size=None, input_len=None,
               embed_dim=128, lstm_units=64, dropout=0.5, optimizer="adam"):
    """
    Función para construir el modelo Bidirectional LSTM.
    Usa sparse_categorical_crossentropy.
    Argumentos:
        vocab_size (int): Tamaño del vocabulario efectivo (incl. padding).
        input_len (int): Longitud de las secuencias de entrada.
        embed_dim (int): Dimensión de los vectores de embedding.
        lstm_units (int): Número de unidades en la capa LSTM (Bidirectional).
        dropout (float): Tasa de dropout.
        optimizer (str): Nombre del optimizador ('adam', 'rmsprop', etc.).
    """
    if vocab_size is None or input_len is None:
         raise ValueError("Los parámetros vocab_size y input_len deben ser proporcionados.")

    model = Sequential([
        # Capa de Embedding
        layers.Embedding(vocab_size, embed_dim, input_length=input_len),
        # Capa LSTM Bidireccional
        layers.Bidirectional(layers.LSTM(lstm_units)),
        # Capa Dropout
        layers.Dropout(dropout),
        # Capa Densa de salida (con num_classes unidades y softmax)
        layers.Dense(num_classes, activation="softmax")
    ])

    # Compilar el modelo
    # Usamos sparse_categorical_crossentropy porque las etiquetas y_train/y_test son enteros (0, 1, 2)
    model.compile(loss="sparse_categorical_crossentropy",
                  optimizer=optimizer, # Puede ser 'adam' o el que se pase (ej. en GridSearchCV)
                  metrics=["accuracy"])

    return model

# --- Funciones de ayuda para Evaluación y Plotting (Adaptadas para etiquetas enteras) ---

def evaluate_model(model, X_test, y_test_labels, target_names, model_name):
    """
    Evalúa un modelo Keras en el conjunto de prueba (con etiquetas enteras) y reporta métricas.
    """
    print(f"\n--- Evaluación de {model_name} en el Conjunto de Prueba ---")
    print("Realizando predicciones en el conjunto de prueba...")
    # predict() en Keras para modelos Sequential con softmax output ya da probabilidades.
    y_pred_proba = model.predict(X_test)

    # Convertir las probabilidades a las etiquetas de clase predichas (índices numéricos enteros)
    y_pred_labels = np.argmax(y_pred_proba, axis=1)

    # y_test_labels ya está en formato de etiquetas numéricas enteras 1D

    # Calcular las métricas de evaluación requeridas
    accuracy = accuracy_score(y_test_labels, y_pred_labels)
    precision = precision_score(y_test_labels, y_pred_labels, average='weighted', zero_division=0)
    recall = recall_score(y_test_labels, y_pred_labels, average='weighted', zero_division=0)
    f1 = f1_score(y_test_labels, y_pred_labels, average='weighted', zero_division=0)
    kappa = cohen_kappa_score(y_test_labels, y_pred_labels)

    print(f"\nMétricas de Evaluación ({model_name}):")
    print(f"Precisión (Accuracy): {accuracy:.4f}")
    print(f"Precisión (Precision - Weighted): {precision:.4f}")
    print(f"Exhaustividad (Recall - Weighted): {recall:.4f}")
    print(f"Puntuación F1 (F1-score - Weighted): {f1:.4f}")
    print(f"Kappa de Cohen: {kappa:.4f}")

    print("\nReporte Detallado de Clasificación:")
    # classification_report espera etiquetas enteras y target_names
    print(classification_report(y_test_labels, y_pred_labels, target_names=target_names, zero_division=0))

    # Graficar la matriz de confusión
    plot_confusion_matrix(y_test_labels, y_pred_labels, target_names, model_name)

    return y_test_labels, y_pred_labels # Retorna las etiquetas numéricas

def plot_history(history, title):
    """
    Grafica el historial de entrenamiento y validación (accuracy y loss).
    """
    if not isinstance(history, History):
        # Esto puede ocurrir si se usa validation_split en el fit dentro de KerasClassifier
        # en GridSearchCV, ya que KerasClassifier puede no retornar un History object estándar.
        # Sin embargo, para el modelo fijo y el modelo retrainado para history, debería funcionar.
        print(f"Advertencia: No se pudo graficar el historial para '{title}' (objeto history no válido).")
        return

    plt.figure(figsize=(12, 4))

    # Plot Accuracy
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train Accuracy')
    if 'val_accuracy' in history.history: # Solo grafica validación si está disponible
        plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title(f'{title} - Accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epochs')
    plt.legend()

    # Plot Loss
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train Loss')
    if 'val_loss' in history.history: # Solo grafica validación si está disponible
        plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title(f'{title} - Loss')
    plt.ylabel('Loss')
    plt.xlabel('Epochs')
    plt.legend()

    plt.tight_layout()
    plt.show()

def plot_confusion_matrix(y_true, y_pred, target_names, title):
    """
    Grafica la matriz de confusión usando Seaborn con anotaciones numéricas.
    """
    conf_matrix = confusion_matrix(y_true, y_pred)

    plt.figure(figsize=(8, 6))
    # Usar los nombres de las clases para las etiquetas de los ejes
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=True, # cbar=True para mostrar la barra de color
                xticklabels=target_names, yticklabels=target_names) # Usar target_names para las etiquetas
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.title(f'{title} - Confusion Matrix')
    plt.show()


# ================================================================
# 1. Experimento 1: LSTM baseline (Usando test_clean)
# ================================================================
print("\n" + "="*60)
print("--- 1. LSTM Modelo con Hiperparámetros Fijos (Baseline - Usando test_clean) ---")
print("="*60)

# Definir hiperparámetros fijos para el modelo baseline
FIXED_LSTM_UNITS = 128 # <-- Usar 64 como en tu código que funcionó
FIXED_DENSE_UNITS = 64 # No hay capa Dense separada en tu función build_lstm
FIXED_DROPOUT_RATE = 0.5 # <-- Usar 0.5 como en tu código
FIXED_LEARNING_RATE = 0.001 # Adam default en tu código
FIXED_EPOCHS = 10 # <-- Usar 10 como en tu código
FIXED_BATCH_SIZE = 256 # <-- Usar 64 como en tu código

print(f"Hiperparámetros Fijos para Evaluación (Usando test_clean):")
print(f"  Modelo: Bidirectional LSTM")
print(f"  LSTM Units: {FIXED_LSTM_UNITS} (Bidirectional)")
print(f"  Dropout Rate: {FIXED_DROPOUT_RATE}")
print(f"  Optimizer: Adam (Default)")
print(f"  Epochs: {FIXED_EPOCHS} (usado para entrenamiento en train+validation_split)")
print(f"  Batch Size: {FIXED_BATCH_SIZE} (usado para entrenamiento en train+validation_split)")
print(f"  Validation Split: {VAL_SPLIT_IN_FIT}")


# Construir el modelo LSTM con los hiperparámetros fijos para entrenamiento de EVALUACIÓN
# Este modelo se entrena solo en X_train (usando validation_split) para poder evaluar en X_test
fixed_lstm_model_eval = build_lstm_model(
    vocab_size=effective_vocab_size,
    input_len=MAX_SEQUENCE_LENGTH,
    embed_dim=EMBEDDING_DIM,
    lstm_units=FIXED_LSTM_UNITS,
    dropout=FIXED_DROPOUT_RATE,
    optimizer='adam' # Fijo en 'adam' como en tu función
)

# Mostrar el resumen del modelo
print("\nResumen del Modelo LSTM Fijo (para Evaluación - Usando test_clean):")
fixed_lstm_model_eval.summary()

# Entrenar el modelo para EVALUACIÓN con los datos de entrenamiento (X_train)
print(f"\nEntrenando el modelo LSTM fijo (para Evaluación - Usando test_clean) por {FIXED_EPOCHS} epochs en X_train (con validation_split={VAL_SPLIT_IN_FIT})...")
fixed_history = fixed_lstm_model_eval.fit(
    X_train,
    y_train, # Etiquetas enteras
    epochs=FIXED_EPOCHS,
    batch_size=FIXED_BATCH_SIZE,
    validation_split=VAL_SPLIT_IN_FIT, # Usar validation_split en fit
    verbose=1 # Mostrar progreso del entrenamiento
)

# Graficar el historial de entrenamiento (usando los datos de train+val)
plot_history(fixed_history, "LSTM Modelo Fijo (Entrenamiento en Train - Usando test_clean)")

# Evaluar el modelo fijo en el conjunto de prueba (X_test, y_test)
evaluate_model(fixed_lstm_model_eval, X_test, y_test, target_names, "LSTM Modelo Fijo (Evaluación en Test - Usando test_clean)")

# --- Entrenamiento Final y Guardado del Modelo LSTM Fijo (Usando TODOS los Datos) ---
print("\n--- Entrenamiento Final y Guardado del Modelo LSTM Fijo (Usando TODOS los Datos - test_clean) ---")
print("Construyendo y entrenando el modelo final fijo en todo el dataset (X_padded, y)...")

# Construir un NUEVO modelo con los mismos hiperparámetros fijos
fixed_lstm_model_final = build_lstm_model(
    vocab_size=effective_vocab_size,
    input_len=MAX_SEQUENCE_LENGTH,
    embed_dim=EMBEDDING_DIM,
    lstm_units=FIXED_LSTM_UNITS,
    dropout=FIXED_DROPOUT_RATE,
    optimizer='adam' # Fijo en 'adam'
)

# Entrenar el modelo FINAL en TODO el dataset (X_padded, y)
# Usamos el mismo número de epochs y batch_size que en el entrenamiento de evaluación.
# No hay validation_data ni validation_split aquí, ya que usamos todos los datos disponibles.
fixed_lstm_model_final.fit(
    X_padded, # Usar todos los datos (derivados de test_clean)
    y,        # Usar todas las etiquetas (enteras)
    epochs=FIXED_EPOCHS,
    batch_size=FIXED_BATCH_SIZE,
    verbose=1 # Mostrar progreso del entrenamiento
)

# Guardar este modelo FINAL entrenado en todos los datos
fixed_model_save_path = os.path.join(models_output_folder, 'lstm_fixed_model_full_data_testclean')
print(f"\nGuardando el Modelo LSTM Fijo FINAL (entrenado en todos los datos) en '{fixed_model_save_path}'...")
try:
    fixed_lstm_model_final.save(fixed_model_save_path)
    print("Modelo fijo FINAL guardado exitosamente.")
except Exception as e:
    print(f"Error al guardar el modelo fijo FINAL: {e}")
    print("El modelo fijo FINAL no pudo ser guardado. Revisa los permisos de escritura o la ruta especificada.")
```

```python
# Notebook: notebooks/LSTM_sentiment_classifier_TestClean_Final.ipynb
print("Init: Starting LSTM Sentiment Classifier Notebook (using test_clean, integrated logic)...")
import random
from tensorflow.keras import layers

# --- Título del Notebook ---
# # Clasificación de Sentimiento de Tweets con LSTM y GridSearchCV (Usando 'test_clean' y Lógica Integrada)

# --- Descripción ---
# Este notebook implementa, entrena y evalúa dos modelos de Red Neuronal LSTM
# para clasificar el sentimiento de tweets (positivo, negativo, neutral)
# del dataset del Mundial FIFA 2022, utilizando aprendizaje supervisado.
# Utiliza la columna 'test_clean' como entrada de texto, Bidirectional LSTM
# y sparse_categorical_crossentropy con etiquetas numéricas enteras.
# Se presenta un modelo LSTM con hiperparámetros fijos como baseline,
# y otro modelo LSTM optimizado usando búsqueda de hiperparámetros con GridSearchCV.
# Incluye preprocesamiento (carga de 'test_clean'), división de datos (train/val/test),
# entrenamiento de modelos para evaluación y para despliegue (usando todos los datos),
# evaluación estadística (métricas y matriz de confusión con anotaciones), visualización
# del historial de entrenamiento (para el modelo fijo) y guardado de ambos modelos finales.

# --- Objetivos Cubiertos (Unidad 3) ---
# - OE3.3: Comprensión de aprendizaje supervisado, PNL y redes neuronales (LSTM Bidireccional).
# - OE3.4: Aplicación de redes neuronales (LSTM) para resolver tareas de PNL (clasificación de sentimiento) usando programación.
# - Cumplimiento del requisito de "Find its best hyperparameters using functions such as GridSearchCV", aprovechando el paralelismo de CPU.
# - Evaluación detallada con métricas (Accuracy, Precision, Recall, F1-score, Kappa) y visualizaciones (historial, matriz de confusión).
# - Guardado de modelos entrenados para su uso futuro.

# --- Importaciones Necesarias ---
import pandas as pd
import numpy as np
import os
# import ast # No necesitamos ast.literal_eval ya que test_clean es string
import sys # Para salir si hay errores críticos
import warnings # Para manejar advertencias si es necesario
import matplotlib.pyplot as plt # Para gráficos
import seaborn as sns # Para gráficos más bonitos (especialmente matriz de confusión)


# Sklearn para split de datos y métricas
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, cohen_kappa_score
from sklearn.metrics import confusion_matrix # Para la matriz de confusión
from sklearn.metrics import classification_report # Para reporte completo de clasificación

# TensorFlow/Keras para la red neuronal
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Bidirectional, Dense, Dropout # Importar Bidirectional
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
# from tensorflow.keras.utils import to_categorical # No necesaria con sparse_categorical_crossentropy y etiquetas enteras
from tensorflow.keras.callbacks import History # Para capturar el historial en el modelo fijo

# Wrapper para usar modelos Keras con GridSearchCV de scikit-learn
from scikeras.wrappers import KerasClassifier

# Para la búsqueda de hiperparámetros en Keras/TensorFlow con Scikit-learn
from sklearn.model_selection import GridSearchCV

# Ignorar warnings de scikit-learn y KerasClassifier que a veces pueden aparecer
warnings.filterwarnings('ignore', category=UserWarning, module='sklearn')
warnings.filterwarnings('ignore', category=FutureWarning, module='sklearn')
warnings.filterwarnings('ignore', category=DeprecationWarning, module='scikeras')
warnings.filterwarnings('ignore', category=UserWarning, module='tensorflow')


# ================================================================
# 0. Silencio de logs de TensorFlow y Semilla para reproducibilidad
# ================================================================
os.environ["TF_CPP_MIN_LOG_LEVEL"]  = "3"   # 0=DEBUG,1=INFO,2=WARNING,3=ERROR
os.environ["TF_ENABLE_ONEDNN_OPTS"] = "0"   # (opcional) desactiva oneDNN

SEED = 42
random.seed(SEED); np.random.seed(SEED); tf.random.set_seed(SEED)

# --- Configuración de Hardware (Forzar uso de CPU) ---
print("Configurando TensorFlow para usar solo CPU.")
try:
    gpu_devices = tf.config.list_physical_devices('GPU')
    if gpu_devices:
        tf.config.set_visible_devices([], 'GPU')
        print(f"GPUs detectadas: {len(gpu_devices)}. Explicitamente deshabilitadas para usar solo CPU.")
    else:
        print("No se detectaron GPUs físicas. TensorFlow usará CPU por defecto.")
    print("Dispositivos físicos visibles para TensorFlow:", tf.config.list_physical_devices())
except Exception as e:
    print(f"Error al intentar configurar TensorFlow para solo CPU: {e}")
    print("Continuando. TensorFlow usará CPU por defecto si la configuración falló o no hay GPU.")


# --- Configuración de Rutas y Parámetros ---
input_folder = '../data_processed' # O la ruta correcta a tu archivo
input_filename = 'fifa_tweets_clean.csv' # Nombre del archivo CSV
input_csv_file = os.path.join(input_folder, input_filename)

# Definir la carpeta de salida para los modelos guardados
ruta_base_proyecto = os.path.dirname(os.getcwd())
models_output_folder = os.path.join(ruta_base_proyecto, 'models')
os.makedirs(models_output_folder, exist_ok=True)
print(f"Los modelos entrenados se guardarán en: {models_output_folder}")


# Definir las columnas a usar del archivo procesado
CLEAN_TEXT_INPUT_COLUMN = 'test_clean' # Columna con el texto limpio (string)
LABEL_COLUMN = 'sentiment_label'     # Columna con la etiqueta numérica del sentimiento (int 0, 1, 2)
ORIGINAL_SENTIMENT_COLUMN = 'Sentiment' # Columna original del sentimiento (para referencia/mapeo inverso)


# Hyperparámetros para la preparación de datos (Tokenización y Padding)
MAX_WORDS = 10000         # Número máximo de palabras únicas en el vocabulario del tokenizer (las más frecuentes). None para usar todo el vocabulario.
MAX_SEQUENCE_LENGTH = 50 # <-- Usar MAX_LEN=50 como en tu código que funcionó
EMBEDDING_DIM = 128       # <-- Usar embed_dim=128 como en tu código que funcionó

# Configuración de entrenamiento y evaluación
TEST_SIZE = 0.20          # Proporción del dataset para el conjunto de prueba (20%)
# VAL_SIZE no se usa en split explícito con validation_split en .fit, pero mantenemos la proporción para el split inicial si se necesitara en GridSearchCV cv
VAL_SPLIT_IN_FIT = 0.1    # Proporción del conjunto de entrenamiento (X_train) a usar para validación DENTRO de .fit()
RANDOM_STATE = SEED       # Para reproducibilidad

# --- Cargar el dataset procesado ---
print(f"\nCargando dataset procesado desde '{input_csv_file}'...")
try:
    df = pd.read_csv(
        input_csv_file,
        encoding='utf-8',
        # No necesitamos converters para 'processed_tokens' si no la usamos para X.
    )

    print("Dataset procesado cargado exitosamente.")
    print(f"Número total de filas cargadas: {len(df)}")
    print(f"Columnas cargadas: {df.columns.tolist()}")

except FileNotFoundError:
    print(f"Error: El archivo procesado '{input_csv_file}' no fue encontrado.")
    print(f"Asegúrate de que la ruta '{input_csv_file}' es correcta y que el archivo existe.")
    sys.exit(1)
except Exception as e:
    print(f"Error al cargar o procesar el archivo CSV: {e}")
    print("Asegúrate de que el archivo no esté corrupto.")
    sys.exit(1)

# --- Verificación y Preparación de Datos ---

# 1. Eliminar filas con etiquetas de sentimiento faltantes o no válidas
initial_rows = len(df)
df_cleaned = df[df[LABEL_COLUMN] != -1].copy()
print(f"\nFilas con sentiment_label = -1 eliminadas: {initial_rows - len(df_cleaned)}")

# 2. Eliminar filas donde la columna de texto de entrada es NaN o está vacía/solo espacios
# Convertir a string antes de verificar NaN por si acaso
df_cleaned[CLEAN_TEXT_INPUT_COLUMN] = df_cleaned[CLEAN_TEXT_INPUT_COLUMN].astype(str)
df_cleaned = df_cleaned[df_cleaned[CLEAN_TEXT_INPUT_COLUMN].apply(lambda x: x.strip() != "")].copy()
print(f"Filas con {CLEAN_TEXT_INPUT_COLUMN} vacío o NaN eliminadas: {initial_rows - len(df_cleaned) - (initial_rows - len(df_cleaned))}") # Corrección en el cálculo

# Convertir las etiquetas numéricas a tipo entero (verificación)
df_cleaned[LABEL_COLUMN] = df_cleaned[LABEL_COLUMN].astype(int)

print(f"\nNúmero final de filas para modelado: {len(df_cleaned)}")
# Verificar la distribución de las etiquetas numéricas válidas
print(f"Distribución de etiquetas ({LABEL_COLUMN}): \n{df_cleaned[LABEL_COLUMN].value_counts()}")

# Determinar el número de clases y el mapeo inverso
unique_labels = sorted(df_cleaned[LABEL_COLUMN].unique())
num_classes = len(unique_labels)
print(f"Número de clases detectado: {num_classes}")

# Mapeo inverso de etiquetas numéricas a nombres
reverse_sentiment_mapping = {0: 'negative', 1: 'neutral', 2: 'positive'}
valid_target_names = [reverse_sentiment_mapping[i] for i in unique_labels if i in reverse_sentiment_mapping]
if len(valid_target_names) != num_classes:
    print("Advertencia: Las etiquetas numéricas en los datos no coinciden completamente con el mapeo reverse_sentiment_mapping.")
    target_names = [f'Class_{i}' for i in unique_labels] # Crea nombres genéricos si el mapeo falla
else:
    target_names = valid_target_names

print(f"Mapeo de etiquetas: {reverse_sentiment_mapping}")
print(f"Nombres de clases para reporte: {target_names}")


# 3. Separar características (X) y etiquetas (y)
X = df_cleaned[CLEAN_TEXT_INPUT_COLUMN] # Serie de strings de 'test_clean'
y = df_cleaned[LABEL_COLUMN] # Serie/array de etiquetas numéricas enteras


# 4. Tokenización y construcción del vocabulario (con Keras Tokenizer)
print(f"\nRealizando tokenización y padding usando '{CLEAN_TEXT_INPUT_COLUMN}'...")
# Usar num_words=MAX_WORDS para limitar el vocabulario
tokenizer = Tokenizer(num_words=MAX_WORDS, oov_token="<OOV>") # oov_token para palabras fuera del vocabulario
# Ajustar el tokenizer DIRECTAMENTE en la Serie X (que contiene las strings de test_clean)
tokenizer.fit_on_texts(X)

# Convertir el texto a secuencias de enteros usando el vocabulario aprendido
X_sequences = tokenizer.texts_to_sequences(X)

# Imprimir tamaño del vocabulario real y efectivo para Embedding
# input_dim en Embedding debe ser 1 + máximo índice entero en los datos de entrada.
# Si num_words=N, los índices más altos serán <= N si no hay OOV, o N+1 si hay OOV.
# len(tokenizer.word_index) da el número de palabras únicas *vistas*, sin padding.
# Si num_words es None, vocab_size = len(word_index) + 1 (para padding).
# Si num_words=N, vocab_size = min(len(word_index), N) + 1 (para padding).
# PERO, si hay OOV, y num_words=N, OOV token gets index 1, top N words get 2..N+1. Padding is 0. Size is N+2.
# La forma más segura: obtener el tamaño del vocabulario del tokenizer + 1 (para padding).
# Si num_words está definido, el tokenizer mapeará solo hasta ese límite (+ OOV).
# El tamaño real de los embeddings necesarios es el máximo índice + 1.
# Keras Tokenizer word_index starts from 1. Max index is len(word_index). Add 1 for padding.
effective_vocab_size = len(tokenizer.word_index) + 1
if MAX_WORDS is not None and MAX_WORDS > 0 and MAX_WORDS < len(tokenizer.word_index):
    # Si num_words limitó el vocabulario, el tokenizer.word_index aún mapea todas las palabras vistas,
    # pero texts_to_sequences solo usa los índices de las top num_words (+OOV).
    # El índice OOV (si existe) suele ser 1. Los top num_words suelen ir de 2 a num_words+1. Padding es 0.
    # Total size es num_words + 2 si hay OOV, o num_words + 1 si no hay OOV.
    # Sin embargo, Scikeras wrapper con sparse_categorical_crossentropy parece esperar input_dim = len(word_index) + 1
    # si num_words is None, o MAX_WORDS + 1 si num_words is set (sin contar OOV index special case).
    # Vamos a usar len(tokenizer.word_index) + 1 que es el tamaño total del vocabulario que el tokenizer podría generar.
    # Si MAX_WORDS está seteado, el tokenizer.word_index contendrá más palabras, pero texts_to_sequences y Embedding respetarán MAX_WORDS.
    # Revertir a la lógica simple de len(tokenizer.word_index) + 1
    pass # effective_vocab_size ya está calculado

print(f"Tamaño del vocabulario (num_words en Tokenizer): {MAX_WORDS} (None para todo)")
print(f"Tamaño real del word_index del Tokenizer: {len(tokenizer.word_index)}")
print(f"Tamaño efectivo del vocabulario a usar en Embedding (incl. padding): {effective_vocab_size}")
print(f"Primeras 10 palabras del vocabulario (índice a palabra):")
index_word = {index: word for word, index in tokenizer.word_index.items()}
for i in range(1, min(11, len(index_word) + 1)):
     if i in index_word:
         print(f"  {i}: {index_word[i]}")


# 5. Padding de secuencias
# Asegurar que todas las secuencias tengan la misma longitud para TODO el dataset
X_padded = pad_sequences(X_sequences, maxlen=MAX_SEQUENCE_LENGTH, padding='post', truncating='post')

# Las etiquetas 'y' ya están como enteros, perfectas para sparse_categorical_crossentropy.
# No necesitamos y_one_hot para el entrenamiento, solo para stratify en train_test_split si sklearn lo requiriera con enteros.
# Para estratificar, sklearn.model_selection.train_test_split funciona bien con etiquetas enteras 1D.

# 6. Dividir los datos en conjuntos de entrenamiento y prueba (etiquetas enteras 1D)
# Nota: Aquí no creamos X_val/y_val explícitamente en el split inicial,
# usamos validation_split dentro del .fit() del modelo fijo y el retrain del mejor modelo.
# GridSearchCV usa cv folds para validación, no necesita X_val/y_val explícito aquí.
X_train, X_test, y_train, y_test = train_test_split(
    X_padded, # Usar X_padded (secuencias numéricas)
    y,        # Usar y (etiquetas enteras 1D)
    test_size=TEST_SIZE,
    random_state=RANDOM_STATE,
    stratify=y # Estratificar usando las etiquetas enteras 1D
)

print(f"\n--- Datos listos para el modelado ---")
print(f"Forma de los datos de entrenamiento (X_train): {X_train.shape}")
print(f"Forma de las etiquetas de entrenamiento (y_train - enteros): {y_train.shape}")
print(f"Forma de los datos de prueba (X_test): {X_test.shape}")
print(f"Forma de las etiquetas de prueba (y_test - enteros): {y_test.shape}")
print(f"Forma de los datos COMPLETOS (X_padded): {X_padded.shape}") # Usado para entrenamiento final
print(f"Forma de las etiquetas COMPLETAS (y - enteros): {y.shape}") # Usado para entrenamiento final


print(f"Longitud máxima de secuencia: {MAX_SEQUENCE_LENGTH}")
print(f"Dimensión de embedding: {EMBEDDING_DIM}")
print(f"Tamaño efectivo del vocabulario usado: {effective_vocab_size}")
print(f"Número de clases: {num_classes}")


# --- Definir la función constructora del modelo LSTM Bidireccional ---
# Usa sparse_categorical_crossentropy
def build_lstm_model(vocab_size=None, input_len=None,
               embed_dim=128, lstm_units=64, dropout=0.5, optimizer="adam"):
    """
    Función para construir el modelo Bidirectional LSTM.
    Usa sparse_categorical_crossentropy.
    Argumentos:
        vocab_size (int): Tamaño del vocabulario efectivo (incl. padding).
        input_len (int): Longitud de las secuencias de entrada.
        embed_dim (int): Dimensión de los vectores de embedding.
        lstm_units (int): Número de unidades en la capa LSTM (Bidirectional).
        dropout (float): Tasa de dropout.
        optimizer (str): Nombre del optimizador ('adam', 'rmsprop', etc.).
    """
    if vocab_size is None or input_len is None:
         raise ValueError("Los parámetros vocab_size y input_len deben ser proporcionados.")

    model = Sequential([
        # Capa de Embedding
        layers.Embedding(vocab_size, embed_dim, input_length=input_len),
        # Capa LSTM Bidireccional
        layers.Bidirectional(layers.LSTM(lstm_units)),
        # Capa Dropout
        layers.Dropout(dropout),
        # Capa Densa de salida (con num_classes unidades y softmax)
        layers.Dense(num_classes, activation="softmax")
    ])

    # Compilar el modelo
    # Usamos sparse_categorical_crossentropy porque las etiquetas y_train/y_test son enteros (0, 1, 2)
    model.compile(loss="sparse_categorical_crossentropy",
                  optimizer=optimizer, # Puede ser 'adam' o el que se pase (ej. en GridSearchCV)
                  metrics=["accuracy"])

    return model

# --- Funciones de ayuda para Evaluación y Plotting (Adaptadas para etiquetas enteras) ---

def evaluate_model(model, X_test, y_test_labels, target_names, model_name):
    """
    Evalúa un modelo Keras en el conjunto de prueba (con etiquetas enteras) y reporta métricas.
    """
    print(f"\n--- Evaluación de {model_name} en el Conjunto de Prueba ---")
    print("Realizando predicciones en el conjunto de prueba...")
    # predict() en Keras para modelos Sequential con softmax output ya da probabilidades.
    y_pred_proba = model.predict(X_test)

    # Convertir las probabilidades a las etiquetas de clase predichas (índices numéricos enteros)
    y_pred_labels = np.argmax(y_pred_proba, axis=1)

    # y_test_labels ya está en formato de etiquetas numéricas enteras 1D

    # Calcular las métricas de evaluación requeridas
    accuracy = accuracy_score(y_test_labels, y_pred_labels)
    precision = precision_score(y_test_labels, y_pred_labels, average='weighted', zero_division=0)
    recall = recall_score(y_test_labels, y_pred_labels, average='weighted', zero_division=0)
    f1 = f1_score(y_test_labels, y_pred_labels, average='weighted', zero_division=0)
    kappa = cohen_kappa_score(y_test_labels, y_pred_labels)

    print(f"\nMétricas de Evaluación ({model_name}):")
    print(f"Precisión (Accuracy): {accuracy:.4f}")
    print(f"Precisión (Precision - Weighted): {precision:.4f}")
    print(f"Exhaustividad (Recall - Weighted): {recall:.4f}")
    print(f"Puntuación F1 (F1-score - Weighted): {f1:.4f}")
    print(f"Kappa de Cohen: {kappa:.4f}")

    print("\nReporte Detallado de Clasificación:")
    # classification_report espera etiquetas enteras y target_names
    print(classification_report(y_test_labels, y_pred_labels, target_names=target_names, zero_division=0))

    # Graficar la matriz de confusión
    plot_confusion_matrix(y_test_labels, y_pred_labels, target_names, model_name)

    return y_test_labels, y_pred_labels # Retorna las etiquetas numéricas

def plot_history(history, title):
    """
    Grafica el historial de entrenamiento y validación (accuracy y loss).
    """
    if not isinstance(history, History):
        # Esto puede ocurrir si se usa validation_split en el fit dentro de KerasClassifier
        # en GridSearchCV, ya que KerasClassifier puede no retornar un History object estándar.
        # Sin embargo, para el modelo fijo y el modelo retrainado para history, debería funcionar.
        print(f"Advertencia: No se pudo graficar el historial para '{title}' (objeto history no válido).")
        return

    plt.figure(figsize=(12, 4))

    # Plot Accuracy
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train Accuracy')
    if 'val_accuracy' in history.history: # Solo grafica validación si está disponible
        plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title(f'{title} - Accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epochs')
    plt.legend()

    # Plot Loss
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train Loss')
    if 'val_loss' in history.history: # Solo grafica validación si está disponible
        plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title(f'{title} - Loss')
    plt.ylabel('Loss')
    plt.xlabel('Epochs')
    plt.legend()

    plt.tight_layout()
    plt.show()

def plot_confusion_matrix(y_true, y_pred, target_names, title):
    """
    Grafica la matriz de confusión usando Seaborn con anotaciones numéricas.
    """
    conf_matrix = confusion_matrix(y_true, y_pred)

    plt.figure(figsize=(8, 6))
    # Usar los nombres de las clases para las etiquetas de los ejes
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=True, # cbar=True para mostrar la barra de color
                xticklabels=target_names, yticklabels=target_names) # Usar target_names para las etiquetas
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.title(f'{title} - Confusion Matrix')
    plt.show()

# ================================================================
# 2. Experimento 2: GridSearchCV + SciKeras (Usando test_clean)
# ================================================================
print("\n" + "="*60)
print("--- 2. LSTM Modelo con GridSearchCV (Optimizado - Usando test_clean) ---")
print("="*60)

# Usaremos los mismos epochs y batch_size para cada trial dentro de GridSearchCV que en el baseline fit
GRID_EPOCHS = 20 # <-- Usar 10 como en tu código
GRID_BATCH_SIZE = 3 # <-- Usar 64 como en tu código

# Crear el wrapper KerasClassifier para GridSearchCV
# Usamos la función build_lstm_model (ya adaptada para sparse_categorical_crossentropy y Bidirectional)
# Nota: KerasClassifier con sparse_categorical_crossentropy espera etiquetas enteras para y en fit()
clf = KerasClassifier(
    model=build_lstm_model,
    # Pasar parámetros fijos necesarios a la función constructora del modelo
    model__vocab_size=effective_vocab_size,
    model__input_len=MAX_SEQUENCE_LENGTH, # Pasar input_len al constructor
    model__embed_dim=EMBEDDING_DIM, # Pasa el valor fijo como default para el grid
    # Parámetros de fit para KerasClassifier (se usarán como defaults si no están en param_grid)
    epochs=GRID_EPOCHS,
    batch_size=GRID_BATCH_SIZE,
    verbose=0 # No imprimir el progreso de entrenamiento de Keras durante la búsqueda
    # No pasar validation_split aquí; GridSearchCV maneja la validación con cv
)

# Definir la cuadrícula de parámetros a buscar (adaptada de tu código)
# embed_dim, lstm_units, dropout, optimizer, epochs, batch_size están en param_grid
param_grid = [
    {
        "model__embed_dim": [128],
        "model__lstm_units": [64],
        "model__dropout": [0.5],
        "optimizer": ["adam"],
        "epochs": [20],
        "batch_size": [64]
    },
    {
        "model__embed_dim": [128],
        "model__lstm_units": [128],
        "model__dropout": [0.3],
        "optimizer": ["rmsprop"],
        "epochs": [20],
        "batch_size": [32]
    },
    {
        "model__embed_dim": [64],
        "model__lstm_units": [64],
        "model__dropout": [0.5],
        "optimizer": ["adam"],
        "epochs": [20],
        "batch_size": [32]
    }
]


# Configurar GridSearchCV para usar X_train, y_train
print(f"\nIniciando búsqueda de hiperparámetros con GridSearchCV en el conjunto de entrenamiento ({X_train.shape[0]} muestras - Usando test_clean)...")
print("Configurando GridSearchCV para aprovechar todos los núcleos del CPU (n_jobs=-1).")
# Usamos X_train y y_train (etiquetas enteras) para la búsqueda.
# GridSearchCV utilizará internamente su validación cruzada en estos datos.
grid = GridSearchCV(
    estimator=clf,
    param_grid=param_grid,
    scoring="accuracy",
    cv=3,
    n_jobs=1,
    verbose=2
)

# Ejecutar la búsqueda en los datos de entrenamiento (X_train, y_train - enteros)
n_combinations = len(param_grid)
print(f"Ejecutando GridSearchCV con {n_combinations} combinaciones y cv={grid.cv} folds (Usando test_clean).")
print(f"El entrenamiento de cada modelo se ejecutará en CPU, y las {n_combinations * grid.cv} combinaciones/folds se paralelizarán usando n_jobs={grid.n_jobs} núcleos.")
print(f"Epochs por trial definidos manualmente en las combinaciones.")



try:
     # grid.fit espera X (padded sequences) e y (etiquetas enteras)
     grid_search_result = grid.fit(X_train, y_train)

except Exception as e:
    print(f"\nError durante la ejecución de GridSearchCV (Usando test_clean): {e}")
    print("Asegúrate de que TensorFlow y Scikeras estén instalados y configurados.")
    print("Considera reducir el tamaño de la cuadrícula (param_grid), epochs, batch_size o n_jobs.")
    print(f"Detalle del error: {e}")
    raise


# --- Mostrar los resultados de GridSearchCV ---
print("\n--- Resultados de la búsqueda de hiperparámetros (GridSearchCV - Usando test_clean) ---")
print(f"Mejor puntuación media de validación cruzada (según métrica '{grid.scoring}'): {grid_search_result.best_score_:.4f}")
print(f"Mejores hiperparámetros encontrados: {grid_search_result.best_params_}")

# Obtener el mejor estimador de KerasClassifier
best_lstm_classifier_eval = grid_search_result.best_estimator_
# Acceder al modelo Keras subyacente del best_estimator_ (entrenado en todo X_train)
# Este modelo ya tiene los pesos finales después del fit en X_train por GridSearchCV
best_lstm_model_eval = best_lstm_classifier_eval.model_

# Mostrar el resumen del mejor modelo (basado en el KerasClassifier retrainado en X_train)
print("\nResumen del Mejor Modelo LSTM (GridSearchCV - Entrenado en Train - Usando test_clean):")
try:
    best_lstm_model_eval.summary()
except Exception as e:
    print(f"Error): {e}")
# --- Re-entrenar el mejor modelo para obtener el historial de entrenamiento (Usando train+validation_split) ---
# GridSearchCV no proporciona un historial de entrenamiento único para el best_estimator_.
# Para obtener un historial similar al del modelo fijo, reconstruimos y entrenamos el mejor modelo
# con los mejores parámetros encontrados, usando validation_split.
print("\n--- Re-entrenando el Mejor Modelo (GridSearchCV) en X_train con validation_split para obtener Historial ---")
# Obtener los mejores hiperparámetros para la construcción y fit
best_params_for_retrain = grid_search_result.best_params_

# Parámetros específicos para el fit del retrain (epochs y batch_size del mejor trial)
retrain_epochs = best_params_for_retrain.get('epochs', GRID_EPOCHS)
retrain_batch_size = best_params_for_retrain.get('batch_size', GRID_BATCH_SIZE)

# Eliminar epochs y batch_size de best_params_for_retrain para pasarlos al build_model
build_params_for_retrain = {k.replace('model__', ''): v for k, v in best_params_for_retrain.items() if 'model__' in k}
build_params_for_retrain['optimizer'] = best_params_for_retrain['optimizer'] # Agregar el optimizer

print(f"Re-entrenando con params: {build_params_for_retrain}, epochs={retrain_epochs}, batch_size={retrain_batch_size}")

# Construir un NUEVO modelo con los mejores hiperparámetros
best_model_for_history = build_lstm_model(
    vocab_size=effective_vocab_size,
    input_len=MAX_SEQUENCE_LENGTH,
    **build_params_for_retrain # Desempaqueta los parámetros del build_model
)

# Entrenar este modelo en X_train con validation_split para capturar el historial
hist_best = best_model_for_history.fit(
    X_train,
    y_train, # Etiquetas enteras
    epochs=retrain_epochs,
    batch_size=retrain_batch_size,
    validation_split=VAL_SPLIT_IN_FIT, # Usar validation_split
    verbose=1 # Mostrar progreso
)

# Graficar el historial de entrenamiento para el mejor modelo
plot_history(hist_best, "LSTM Mejor Modelo (GridSearchCV) – Training history")


# Evaluar el mejor modelo (entrenado en X_train por GridSearchCV) en el conjunto de prueba (X_test)
# Usamos best_lstm_model_eval que es el .model del best_estimator_ de GridSearchCV
evaluate_model(best_lstm_model_eval, X_test, y_test, target_names, "LSTM Mejor Modelo (Evaluación en Test - Usando test_clean)")


# --- Entrenamiento Final y Guardado del Mejor Modelo (Usando TODOS los Datos) ---
print("\n--- Entrenamiento Final y Guardado del Mejor Modelo (GridSearchCV - Usando TODOS los Datos - test_clean) ---")
print("Construyendo y entrenando el mejor modelo FINAL en todo el dataset (X_padded, y)...")

# Obtener los mejores hiperparámetros para el entrenamiento final (incluyendo epochs y batch_size)
final_best_params = grid_search_result.best_params_
final_build_params = {k.replace('model__', ''): v for k, v in final_best_params.items() if 'model__' in k}
final_build_params['optimizer'] = final_best_params['optimizer']

final_epochs_all_data = final_best_params.get('epochs', GRID_EPOCHS)
final_batch_size_all_data = final_best_params.get('batch_size', GRID_BATCH_SIZE)


print(f"Entrenando modelo final con los mejores hiperparámetros en TODOS los datos ({X_padded.shape[0]} muestras):")
print(f"  Build params: {final_build_params}")
print(f"  Epochs: {final_epochs_all_data}, Batch Size: {final_batch_size_all_data}")

# Construir un NUEVO modelo con los mejores hiperparámetros
best_lstm_model_final = build_lstm_model(
    vocab_size=effective_vocab_size,
    input_len=MAX_SEQUENCE_LENGTH,
    **final_build_params # Desempaqueta los parámetros del build_model
)

# Entrenar el modelo FINAL en TODO el dataset (X_padded, y)
best_lstm_model_final.fit(
    X_padded, # Usar todos los datos (derivados de test_clean)
    y,        # Usar todas las etiquetas (enteras)
    epochs=final_epochs_all_data,
    batch_size=final_batch_size_all_data,
    verbose=1 # Mostrar progreso del entrenamiento final
)

# Guardar este modelo FINAL entrenado en todos los datos
gridsearch_model_save_path = os.path.join(models_output_folder, 'best_lstm_gridsearch_model_full_data_testclean')
print(f"\nGuardando el Mejor Modelo LSTM FINAL (GridSearchCV - entrenado en todos los datos) en '{gridsearch_model_save_path}'...")
try:
    best_lstm_model_final.save(gridsearch_model_save_path)
    print("Mejor modelo FINAL (GridSearchCV) guardado exitosamente.")
except Exception as e:
    print(f"Error al guardar el mejor modelo FINAL (GridSearchCV): {e}")
    print("El mejor modelo FINAL (GridSearchCV) no pudo ser guardado. Revisa los permisos de escritura o la ruta especificada.")


print("\n" + "="*60)
print("--- Implementación y Evaluación de Modelos LSTM Finalizada (Usando test_clean) ---")
print("="*60)

print("\nRelación de este Notebook con los Objetivos de la Unidad 3:")
print(f"- OE3.3 (Comprender ML, PNL, Supervisado, Redes Neuronales): Este notebook aplica los conceptos de aprendizaje supervisado para una tarea de Procesamiento de Lenguaje Natural (clasificación de sentimiento) utilizando una arquitectura de red neuronal (LSTM Bidireccional).")
print(f"- OE3.4 (Aplicar RNN, LSTM, PNL usando programación): Se construye, entrena y evalúa modelos LSTM utilizando las librerías TensorFlow/Keras y scikit-learn para resolver la tarea de clasificación de sentimiento sobre datos de texto preprocesado ('test_clean').")
print(f"- La implementación de un modelo fijo y otro con GridSearchCV permite comparar enfoques y evaluar el impacto de la optimización de hiperparámetros.")
print(f"- El uso de GridSearchCV ('Find its best hyperparameters') y n_jobs=-1 para aprovechar el CPU cumple directamente con el requisito.")
print(f"- La preparación de datos (Tokenización por Keras, Padding) demuestra la aplicación práctica de técnicas necesarias para alimentar datos de texto a modelos neuronales.")
print(f"- La evaluación exhaustiva con métricas como Accuracy, Precision, Recall, F1-score y Kappa, junto con las visualizaciones del historial de entrenamiento (modelo fijo y mejor modelo retrainado) y las matrices de confusión (ambos modelos en test), cumplen con los criterios de evaluación del proyecto y proporcionan el análisis estadístico solicitado.")
print(f"- El guardado de ambos modelos entrenados en el dataset completo permite su reutilización posterior.")

print("\nNotebook execution finished.")
```
