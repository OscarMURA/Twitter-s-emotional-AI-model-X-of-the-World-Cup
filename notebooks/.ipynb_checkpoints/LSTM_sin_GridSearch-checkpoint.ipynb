```python
# Notebook: LSTM_sentiment_classifier_sin_GridSearch.ipynb

print("Init: Starting LSTM Sentiment Classifier (versi√≥n sin GridSearch)...")

import random, os, sys, warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import tensorflow as tf
from tensorflow.keras import layers, Sequential
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.callbacks import History

from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, cohen_kappa_score, confusion_matrix,
    classification_report
)

# --- 0. Configuraci√≥n general y semilla ---
os.environ["TF_CPP_MIN_LOG_LEVEL"] = "3"
SEED = 42
random.seed(SEED); np.random.seed(SEED); tf.random.set_seed(SEED)

# Forzar CPU
try:
    gpus = tf.config.list_physical_devices('GPU')
    tf.config.set_visible_devices([], 'GPU')
except:
    pass

# --- 1. Carga y limpieza de datos ---
INPUT_CSV = '../data_processed/fifa_tweets_clean.csv'
df = pd.read_csv(INPUT_CSV, encoding='utf-8')
df = df[df['sentiment_label'] != -1].copy()
df['test_clean'] = df['test_clean'].astype(str).str.strip()
df = df[df['test_clean'] != ""].copy()
df['sentiment_label'] = df['sentiment_label'].astype(int)

print(f"Filas finales: {len(df)}")
print(df['sentiment_label'].value_counts())

# Mapeo de clases
reverse_map = {0:'negative',1:'neutral',2:'positive'}
target_names = [ reverse_map[i] for i in sorted(df['sentiment_label'].unique()) ]

# --- 2. Tokenizaci√≥n y split ---
MAX_WORDS = 10000
MAX_LEN   =  50
EMBED_DIM = 128

tokenizer = Tokenizer(num_words=MAX_WORDS, oov_token="<OOV>")
tokenizer.fit_on_texts(df['test_clean'])
sequences = tokenizer.texts_to_sequences(df['test_clean'])
X_padded = pad_sequences(sequences, maxlen=MAX_LEN, padding='post', truncating='post')
y = df['sentiment_label'].values

X_train, X_test, y_train, y_test = train_test_split(
    X_padded, y, test_size=0.2, random_state=SEED, stratify=y
)

vocab_size = len(tokenizer.word_index) + 1
print(f"Train shape: {X_train.shape}, Test shape: {X_test.shape}")
print(f"Vocab size: {vocab_size}, Classes: {len(target_names)}")

# --- 3. Definici√≥n del modelo LSTM bidireccional ---
def build_lstm_model(vocab_size, input_len,
                     embed_dim=EMBED_DIM,
                     lstm_units=64,
                     dropout=0.5,
                     optimizer='adam'):
    model = Sequential([
        layers.Embedding(vocab_size, embed_dim, input_length=input_len),
        layers.Bidirectional(layers.LSTM(lstm_units)),
        layers.Dropout(dropout),
        layers.Dense(len(target_names), activation='softmax')
    ])
    model.compile(
        loss='sparse_categorical_crossentropy',
        optimizer=optimizer,
        metrics=['accuracy']
    )
    return model

# --- 4. Entrenamiento y evaluaci√≥n del modelo fijo ---
EPOCHS = 20
BATCH_SIZE = 64
VAL_SPLIT = 0.1

model = build_lstm_model(vocab_size, MAX_LEN)
history = model.fit(
    X_train, y_train,
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    validation_split=VAL_SPLIT,
    verbose=1
)

# Funciones de plotting y evaluaci√≥n
def plot_history(hist, title):
    plt.figure(figsize=(12,4))
    plt.subplot(1,2,1)
    plt.plot(hist.history['accuracy'], label='train')
    plt.plot(hist.history.get('val_accuracy',[]), label='val')
    plt.title(f'{title} ‚Äì Accuracy'); plt.xlabel('Epoch'); plt.legend()
    plt.subplot(1,2,2)
    plt.plot(hist.history['loss'], label='train')
    plt.plot(hist.history.get('val_loss',[]), label='val')
    plt.title(f'{title} ‚Äì Loss'); plt.xlabel('Epoch'); plt.legend()
    plt.tight_layout(); plt.show()

def evaluate_model(m, X, y, names):
    y_proba = m.predict(X)
    y_pred = np.argmax(y_proba, axis=1)
    acc  = accuracy_score(y, y_pred)
    prec = precision_score(y, y_pred, average='weighted', zero_division=0)
    rec  = recall_score(y, y_pred, average='weighted', zero_division=0)
    f1   = f1_score(y, y_pred, average='weighted', zero_division=0)
    kap  = cohen_kappa_score(y, y_pred)
    print(f"Accuracy: {acc:.4f}   Precision: {prec:.4f}")
    print(f"Recall:   {rec:.4f}   F1-score:  {f1:.4f}   Kappa: {kap:.4f}\n")
    print(classification_report(y, y_pred, target_names=names, zero_division=0))
    cm = confusion_matrix(y, y_pred)
    plt.figure(figsize=(6,5))
    sns.heatmap(cm, annot=True, fmt='d',
                xticklabels=names, yticklabels=names, cmap='Blues')
    plt.xlabel('Pred'); plt.ylabel('True'); plt.title('Confusion Matrix'); plt.show()

plot_history(history, "Baseline LSTM")
evaluate_model(model, X_test, y_test, target_names)

# --- 5. Entrenamiento final con todos los datos y guardado ---
final_model = build_lstm_model(vocab_size, MAX_LEN)
final_model.fit(
    X_padded, y,
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    verbose=1
)

out_dir = os.path.join(os.getcwd(), 'models')
os.makedirs(out_dir, exist_ok=True)
save_path = os.path.join(out_dir, 'lstm_sentiment_final')
final_model.save(save_path)
print(f"Modelo final guardado en: {save_path}")
```
